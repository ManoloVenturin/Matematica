Una sfera, di raggio $r$ fissato, è inscritta nel cono $S$ di volume minimo. Qual è la distanza del vertice del cono dalla superficie della sfera?

```python
# Toolbox simbolico
from sympy import *
```

```python
b,y,R = symbols('b,y,R', real=True, positive=True)
```

```python
# b raggop base del cono
# R raggio della sfera
h = y+R # Altezza del cono
h
```

```python
# Volume del cono con h sostituito
V = pi * b**2*h/3
display(V)
```

 I triangoli ABC e AED sono simili

```python
eq = Eq(b / R, h/sqrt(y**2-R**2))
eq
```

```python
bsol = solve(eq,b)[0]
bsol
```

```python
# Sostituisco b nella funzione obiettivo (volume)
V = V.subs(b,bsol) 
V = factor(V)
V
```

```python
# Derivata prima 
V_prime = diff(V,y)
V_prime = factor(V_prime)
V_prime
```

```python
# Ricerca punti di stazionarietà
ysol = solve(V_prime,y)
ysol
```

```python
# Solo quella positiva è accettabile
ypos = 3*R
ypos
```

```python
# Derivata seconda per dimostrare ottimo
V_second = diff(V,y)
V_second = factor(V_second)
V_second
```

```python
# Sostituisco punto nella derivata seconda
V_second.subs(y,ypos)
```

```python
# Bisogna fare lo studio della derivata prima
```

```python
# Base
bsol.subs(y,ypos)
```

```python
# altezza del cono
h.subs(y,ypos)
```

```python
# distanza richiesta
ypos-R
```
