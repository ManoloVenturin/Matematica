√à noto che:
- $ùëÉ''(ùë•) = 12ùë•^2 ‚àí 24ùë•$
- hanno entrambe nell'origine degli assi un flesso a tangente orizzontale
- il valore massimo assunto dalla funzione $ùëÑ$ √® uguale a ${27\over 4}$

Determinare l'espressione analitica delle funzioni $P(x)$ e $Q(x)$

```python
# Toolbox simbolico
from sympy import *
```

```python
x, P, Q = symbols('x, P, Q', real=True)
k, c1, c2 = symbols('k, c1, c2', real=True)
```

```python
P2 = 12 * x**2 -24*x
P2
```

```python
P1 = integrate(P2, x) + c1
P1
```

```python
P = integrate(P1, x) + c2
P
```

```python
# Funzione Q
Q = k * P
```

Le derivate di $Q$ sono quelle di $P$ moltiplicate per $k$

```python
# Equazione: hanno entrambe in x=0 un flesso a tangenza orizzontale
# Il comportamento di Q √® uguale al comportamento di P
# Flesso di P
eq1 = Eq(P.subs(x,0),0)
print(eq1)

eq2 = Eq(P1.subs(x,0),0)
print(eq2)
```

```python
P = P.subs([(c1,0),(c2,0)])
P = factor(P)
P1 = factor(diff(P, x))
print(P)
print(P1)
```

```python
Q = k * P
print(Q)
```

```python
# Valore massimo di Q √® 27/4
# studio della sua derivata prima i.e. P1
Q_prime = factor(k*P1)
Q_prime
```

```python
# Zeri
solve(Q_prime,x)
```

Lo studio della derivata prima, dipende da $k$

Se $k<0$ allora lo studio della derivata prima risulta

```python
p1 = plot(sign(Q_prime.subs(k,-1)),(x,-5,10), ylim=(-2, 2))
```

```python
p1 = plot((Q.subs(k,-1)),(x,-5,10), ylim=(-20, 30))
```

quindi il punto di massimo si ha in $x=3$ da cui l'equazione

```python
eq3 = Eq(k*P.subs(x,3),simplify(27)/4)
eq3
```

da cui la soluzione

```python
ksol = solve(eq3, k)[0]
ksol
```
