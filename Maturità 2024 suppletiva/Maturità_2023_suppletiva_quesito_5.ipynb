Determinare il valore del parametro reale $ùëò$ in modo che la retta di equazione cartesiana $ùë¶ = ùë• ‚àí 2$ risulti tangente alla curva $ùë¶ = ùë•^3 + ùëòùë•$

```python
# Toolbox simbolico
from sympy import *
```

```python
x,y,k,r = symbols('x,y,k,r', real=True)
```

```python
# Equazione della retta
r = x-2
r
```

```python
# Curva parametrica
yk = x**3 + k*x
yk
```

Condizioni di tangenza

- Stesso punto
- Stessa pendenza nel punto di intersezione

```python
# Equazione 1 stesso punto
eq1 = Eq(r,yk)
print_latex(eq1)
eq1
```

```python
# Calcolo derivate
# Retta
m = diff(r,x)
m
```

```python
# yk
yk_prime = diff(yk,x)
yk_prime
```

```python
# Equazione 2: stessa pendenza
eq2 = Eq(m,yk_prime)
print_latex(eq2)
eq2
```

```python
# Soluzione sistema
xksol = solve([eq1,eq2],[x,k])
xksol
```

```python
xsol = xksol[0][0]
ksol = xksol[0][1]
(xsol, ksol)
```

```python
# y di intersezioen
ysol = r.subs(x,xsol)
ysol
```

```python
# Curva yk
y = yk.subs(k,ksol)
y
```

```python
# Visualizzazione
p1 = plot(y,(x,-5,5), ylim=(-10, 10), show=False, markers=[{'args': [[xsol], [ysol], 'ro']}])
p2 = plot(r,(x,-5,5), show=False)
p1.append(p2[0])
p1.show()
```
