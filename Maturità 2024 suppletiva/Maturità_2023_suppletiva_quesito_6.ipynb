Scrivere una funzione polinomiale $ùë¶ = ùëù(ùë•)$ di terzo grado che si annulli solo per $ùë• = 0$ e per $ùë• = 3$, il cui grafico sia tangente all'asse $x$ in un punto e passi per $ùëÉ(1, ‚àí4)$. 

Determinare l'area della regione piana limitata compresa tra l'asse $x$ ed il grafico della funzione polinomiale individuata. 

```python
# Toolbox simbolico
from sympy import *
```

```python
x,a,b,y = symbols('x,a,b,y', real=True)
```

La funzione di terzo grado deve avere una radice doppia o in $x=0$ o in $x=3$ cio√® essere della forma

$$
p(x) = a  x^2 (x-3)
$$

oppure

$$
p(x) = a  x (x-3)^2
$$

Infatti il testo ci dice "il cui grafico sia tangente all'asse  ùë• in un punto"


Utilizziamo l'espressione generale

$$
p(x) = a x (x-3) (x-b) 
$$

dove $b$ pu√≤ assumere solo i valori

- $b=0$, oppure
- $b=3$

```python
p = a*x*(x-3)*(x-b)
display(p)
display(expand(p))
print_latex(expand(p))
```

```python
# Condizione di passaggio per il punto dato
eq1 = Eq(p.subs(x,1),-4)
eq1
```

```python
# Derivata prima
p_prime = expand(diff(p,x))
print_latex(p_prime)
p_prime
```

### Caso 1

```python
# Valutiamo la derivata prima per b=0
bv = 0
p_prime_1 = p_prime.subs(b,bv)
display(p_prime_1)
display(factor(p_prime_1))
```

Si annulla in $x=0$ e $x=2$

Il punto $x=0$ sta nell'asse $x$

```python
xsol_1 = solve(p_prime_1,x)
display(xsol_1)
for xv in xsol_1:
    print(xv, p.subs([(x,xv),(b,bv)]), p_prime_1.subs(x,xv))
```

```python
# Valore di a
eq1_1 = eq1.subs(b,bv)
display(eq1_1)
asol = solve(eq1_1,a)[0]
asol
```

Questa soluzione √® compatibile

```python
p_1 = p.subs([(a,asol),(b,bv)])
p_1
```

```python
p1 = plot(p_1,(x,-5,5), ylim=(-10, 10), markers=[{'args': [[0,3,1], [0,0,-4], 'ro']}])
```

```python
p1 = plot(diff(p_1,x),(x,-5,5), ylim=(-10, 10), markers=[{'args': [[0,2], [0,0], 'ro']}])
```

### Caso 2

```python
# Valutiamo la derivata prima per b=3
bv = 3
p_prime_2 = p_prime.subs(b,bv)
display(p_prime_2)
display(factor(p_prime_2))
```

Si annulla in $x=3$ e $x=1$

Il punto $x=3$ sta nell'asse $x$

```python
xsol_2 = solve(p_prime_2,x)
for xv in xsol_2:
    print(xv, p.subs([(x,xv),(b,bv)]), p_prime_2.subs(x,xv))
```

```python
# Valore di a
eq1_2 = eq1.subs(b,bv)
display(eq1_2)
asol = solve(eq1_2,a)[0]
asol
```

Questa soluzione √® compatibile

```python
p_2 = p.subs([(a,asol),(b,bv)])
p_2
```

```python
p1 = plot(p_2,(x,-5,5), ylim=(-10, 10), markers=[{'args': [[0,3,1], [0,0,-4], 'ro']}])
```

```python
p1 = plot(diff(p_2,x),(x,-5,5), ylim=(-10, 10), markers=[{'args': [[3,1], [0,0], 'ro']}])
```

## Nota
La differenza tra il caso 1 e 2 sta nel fatto che il caso 2 il punto di passaggio √® un punto di minimo relativo
