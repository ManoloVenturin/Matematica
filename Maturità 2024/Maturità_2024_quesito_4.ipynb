## Soluzione come intersezione di due sistemi

Da $x^3+x = \cos{x}$ possiamo scrivere

$$\begin{cases}
y = x^3+x \\
y = \cos{x}
\end{cases}$$


```python
p1 = plot(x**3+x,(x,-3,3), ylim=(-2, 2), show=False, markers=[{'args': [[0.6035196308845124],[0.8233431738830888], 'ro']}])
p2 = plot(cos(x),(x,-3,3), ylim=(-2, 2), show=False)
p1.append(p2[0])
p1.show()
```

```python
# Ricerca dello zero per via numerica
import math
from scipy.optimize import fsolve
```

```python
def func(x):
    return x**3+x-math.cos(x)
xsol = fsolve(func,0.5)[0]
ysol = func(xsol)
ysol_cos = math.cos(xsol)
print(xsol,ysol, ysol_cos)
```
