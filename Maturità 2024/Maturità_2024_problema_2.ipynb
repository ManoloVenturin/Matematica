Si consideri la famiglia di funzioni $f_n(x)=\sqrt[n]{x^2} - \sqrt{ax^2+bx+1}$ con $n\in\mathbb{N}$, $n>1$ e $a,\ b\in\mathbb{R}$, $a<0$.

Il termine $ax^2+bx+1$ rappresenta una parabola verso il basso con condizione di esistenza: $ax^2+bx+1\ge 0$

**Nota:** intero $n>1$ significa da $2$ in poi!

```python
x,y= symbols('x,y', real=True)
a,b = symbols('a,b', real=True)
n = symbols('n', integer=True)
```

```python
f_n = (x**2)**(1/n) -sqrt(a*x**2+b*x+1)
display(f_n)
```

L'espressione √® formata da due termini:il primo non √® simmetrico il secondo s√¨

```python
# Termine 1
t1 = f_n.args[0]
display(t1)

t1.subs([(x,-x)])
display(t1)
```

Quindi scriviamo separatamente i due termini da analizzare

-Il primo termini lo studiere solo per $x\ge0$

```python
f_n_1 = x**(2/n)
display(f_n_1)
```

Il secondo termine lo studieremo normalmente

```python
f_n_2 = -sqrt(a*x**2+b*x+1)
display(f_n_2)
```

## Punto a

Verificare che, qualunque sia il valore di $n$, la funzione $f_n$ non √® derivabile nel punto di ascissa x=0.
Determinare il valore di $n$ in corrispondenza del quale il grafico di $f_n$ presenta un punto angoloso.

```python
# Termine 1
f_n_1_prime = simplify(diff(f_n_1, x))
display(f_n_1_prime)
```

Si tratta di funzione del tipo: 
L'esponente √® $\kappa={2-n\over n} = -\left(1-{2\over n}\right)$ dall'ipotesi $n\ge2$ si ha

1. se $n=2$ allora $\kappa=0$ si tratta del termine $|x|$
2. se $n>2$ allora $-1 < \kappa < 0$ si tratta del termine ${1\over x^{-\kappa}}$ con $0 < -\kappa < 1$

Nel primo caso si tratta della funzione $|x|$ che presenta un punto angoloso

Nel secondo caso la funzione ${1\over x^{-\kappa}}$ con $0 < -\kappa < 1$ ha derivata infinita in 0 e quindi non √® derivabile

```python
k = symbols('k', real=True, positive=True)
limit(1/x**k,x,0, dir="+")
```

Quindi la funzione non √® derivabile
