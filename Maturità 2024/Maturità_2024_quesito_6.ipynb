```python
# Toolbox simbolico
from sympy import *
```

```python
x, t= symbols('x, t', real=True)
n = symbols('n', integer=True)
a = symbols('a', real=True)
```

```python
f = cos(1/t)/t**2
f
```

```python
integrate(f,t)
```

```python
F = integrate(f,(t,a,x))
F
```

```python
diff(F,x)
```

```python
p1 = plot(F.subs(a,1),(x,1,10), ylim=(-5, 5), show=False)
for av in [2,3,4]:
    p2 = plot(F.subs(a,av),(x,av,10), ylim=(-5, 5), show=False)
    p1.append(p2[0])
p1.show()
```

```python
Fterm = F.subs(x,2/pi)
display(Fterm)
Feq = Eq(Fterm,-simplify("1/2"))
Feq
```

```python
asol = solve(Feq, a)
asol
```

```python
sols = solveset(Feq, a, domain=S.Reals)
sols
```

Due famiglie di soluzioni

Per $n=0$ si ha $${6\over \pi}$$
Per $n=1$ si ha $${6\over 5\pi}$$

Il pi√π grande valore si ottiene quando $n=0$ ma non va bene per il vincolo della $x$

Quindi il successivo √® $n=1$


$$
{1\over a} = {\pi\over 6} + k2\pi
$$

$$
{1\over a} = {5\pi\over 6} + k2\pi
$$

per $k=0$ si ha

$$
a = {6\over \pi} > {2\over \pi} \quad\text{ impossibile}
$$

$$
a = {6\over 5\pi} < {2\over \pi}
$$



```python
sol1 = solve(Eq(1/a,pi/6+n*2*pi),a)[0]
print_latex(sol1)
sol1
```

```python
sol2 = solve(Eq(1/a,5*pi/6+n*2*pi),a)[0]
print_latex(sol2)
sol2
```

```python
f1 = lambdify(n, sol1, 'numpy')
f2 = lambdify(n, sol2, 'numpy')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
neval = np.linspace(-10,10,21)
neval
```

```python
plt.figure(figsize=(8,8))
plt.plot(neval,f1(neval), 'ro')
plt.plot(neval,f2(neval), 'bo')
plt.xlabel("n", fontsize=30)
plt.ylabel("a", fontsize=30)
plt.axhline(y=2/np.pi)
plt.axhline(y=0)
plt.text(5,.7,"Vincolo sup", fontsize=20)
plt.text(5,-.1,"Vincolo inf", fontsize=20)
plt.show()
```
