Si consideri la funzione $$f_{a,b}={a x^3+b\over x^2}$$ con $a,\ b \in \mathbb{R}$ 

## Premessa

```python
x,y= symbols('x,y', real=True)
a,b,m,q,k = symbols('a,b,m,q,k', real=True)

f_a_b = (a * x**3 + b) / (x**2)
display(f_a_b)
display(expand(f_a_b))
```

```python
# derivata prima
f_a_b_prime  = diff(f_a_b,x)
display(f_a_b_prime)
display(expand(f_a_b_prime))
```

## Punto a

Determinare i valori dei parametri in modo tale che la retta $t$ di equazione $7x+y-12=0$, sia tangente al grafico di $f_{a,b}(x)$ nel punto $P$ di ascissa $x=1$

```python
# Equazione della retta
t_implicita = Eq(7*x+y,12)
display(f"Retta in forma implicita: {t_implicita}")

# Forma esplicita
t_esplicita = solve(t_implicita, y)[0]
display(f"Retta in forma esplicita: {t_esplicita}")

# Coeffiente angolare
m_retta = -7
display(f"Coefficiente angolare della retta: {m_retta}")

# Valore della retta nel punto x=1
y_retta_1 = t_esplicita.subs([(x,1)])
display(f"y retta in x=1: {y_retta_1}")
```

```python
# Valore della funzione in x=1
yval = f_a_b.subs([(x,1)])
yval
```

```python
# Derivata prima nel punto x=1
f_a_b_prime_in_1 = f_a_b_prime.subs([(x,1)])
display(f"f'(x; a,b) in x=1: {f_a_b_prime_in_1}")
```

```python
# Equazioni di tangenza
# Passaggio per il punto della retta
# Stesso coefficiente angolare

eq1 = Eq(yval, y_retta_1)
print(f"Eq1: {eq1}")

eq2= Eq(f_a_b_prime_in_1, m_retta)
print(f"Eq2: {eq2}")

# Risoluzione del sistema lineare
abtan = solve([eq1, eq2],[a,b])
print(f"Valori di tangenza: {abtan}")
```

### Facciamo il disegno

```python
a_tan = abtan[a]
b_tan = abtan[b]
print(a_tan, b_tan)
```

```python
# Eseguiamo un disegnodi prova: a=1, b=1
f_to_plot = f_a_b.subs([(a,a_tan), (b,b_tan)])
p1 = plot(f_to_plot,(x,-3,3), ylim=(-20, 20), show=False, markers=[{'args': [1, 5, 'ro']}])
p2 = plot(t_esplicita,(x,-3,3), ylim=(-20, 20), show=False)
p1.append(p2[0])
p1.show()
```

## Punto d

Sia $S(k)$, con $k>{3\over 2}$, l'area della regione finita di piano compresa tra la curva $\gamma$, il suo asintoto obliquo, la retta $t$ e la retta di equazione $x=k$.

Calcolare il $\lim\limits_{k\to+\infty} S(k)$, fornendo un'interpretazione geometrica del risultato ottenuto.

```python
# Funzione
display(f)
```

```python
# Retta in forma esplicita
display(t_esplicita)
```

```python
# Equazione asintoto obliquo
y_obliquo = x
display(y_obliquo)
```

```python
# K vincolo
k_vincolo = Eq(x,3/2)
print(k_vincolo)
```

```python
# Visualizzazoine
ymin = -15
ymax = 15
xmax = 5
p1 = plot(f,(x,0,xmax), ylim=(ymin, ymax), show=False)
p2 = plot(t_esplicita,(x,0,xmax), ylim=(ymin, ymax), show=False)
p3 = plot(y_obliquo,(x,0,xmax), ylim=(ymin, ymax), show=False)
p4 = plot_implicit(k_vincolo, (x,0,xmax), ylim=(-ymin, ymax), show=False)
p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
p1.show()
```

```python
# Intersezione retta tangente asintoto
eq = Eq(t_esplicita,y_obliquo)

solve(eq,x)
```

E' proprio il nostro vincolo su $S(k)$ 

L'area √® data da due contributi

- il primo fino a $3/2$ in cui sono coinvoli la fuznione e la retta tangete; fisso indipendente dalla procedura di limite
- il secondo variabile da $3/2$ a $k$ soggetto alla procedura di limite

```python
# Calcolo del primo controbuto dal punto di tangenza a 3/2

# Primitiva
f_integranda1 = expand(f - t_esplicita)
print(f"f integranda1: {f_integranda1}")

f_primitiva1 = integrate(f_integranda1, x)
display(f_primitiva1)

S1 = integrate(f_integranda1, (x, 1, sympify("3/2")))
display(S1)
```

```python
# Calcolo del secondo controbuto dal punto di di intersezioen 3/2 alla retta vertical k

# Primitiva
f_integranda2 = expand(f - y_obliquo)
print(f"f integranda2: {f_integranda2}")

f_primitiva2 = integrate(f_integranda2, x)
display(f_primitiva2)

f_int2 = integrate(f_integranda2, (x, sympify("3/2"), k))
display(f_int2)
```

```python
# Limite del contributo
S2 = limit(f_int2,k,+oo)
display(S2)
```
