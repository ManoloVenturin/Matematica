```python
# Toolbox simbolico
from sympy import *
```

```python
p = symbols('p', real=True)
```

```python
# Problabilit√† evento elementare
P_evento = p **2 * (1-p) **3
P_evento
```

```python
# Numero di eventi elementari
N_eventi = binomial(5,2)
N_eventi
```

```python
# Probabilit√† totale di uscita
P_totale = P_evento * N_eventi
P_totale
```

```python
# Derivata prima
P_prime_totale = factor(diff(P_totale, p))
P_prime_totale
```

```python
# Zeri della derivata prima
pzero = solve(P_prime_totale)
pzero
```

```python
p1 = plot(P_totale,(p,0,1), ylim=(-5, 5), title="P")
```

```python
p1 = plot(sign(P_prime_totale),(p,0,1), ylim=(-5, 5), title="Segno derivata")
```
