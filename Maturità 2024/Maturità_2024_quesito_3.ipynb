```python
# Toolbox simbolico
from sympy import *
```

```python
t,r = symbols('t,r', real=True)
x,y,z = symbols('x,y,z', real=True)
```

## Proeizione del punto nel piano

```python
# Punti A e B
eq_piano = Eq(3*x-2*y+5,0)
display(eq_piano)
P = Array([4,2,1])
P
```

```python
# Coseni direttori del piano
n = Array([3,-2,0])
print_latex(n)
n
```

```python
# Scrittura della retta
r = Eq(Array([x,y,z]), P + t*n)
print_latex(r)
r
```

```python
# equazioni da risolvere
eq1 = eq_piano.lhs
display(eq1)

eq2 = r.lhs[0] - r.rhs[0]
display(eq2)

eq3 = r.lhs[1] - r.rhs[1]
display(eq3)

eq4 = r.lhs[2] - r.rhs[2]
display(eq4)
```

```python
# Trovo t di intersezione
tsol = solve([eq1, eq2, eq3, eq4],[x,y,z,t])
tsol
```
