Il sole sta nel fuoco $F_2$

Notazioni:

- Perielio: $x=\overline{F_2B}$
- Afelio: $y=\overline{F_2A}$ 

Equazione dell'ellisse da determinare:

$$
{x^2\over a^2} + {y^2\over b^2} = 1
$$

```python
# Toolbox simbolico
from sympy import *
```

```python
x, y= symbols('x, y', real=True)
a,b,c,d,k = symbols('a,b,c,d,k', real=True)
```

La somma dell'afelio con il perielio fa due l'asse dell'ellisse, i.e.

$$
2a = x+y
$$

```python
aeq = Eq(2*a, x+y)
print(aeq)
asol = solve(aeq, a)[0]
a = asol
print_latex(a)
a
```

La distanza $c$ √® data da

$$c=a-x$$

```python
c = a-x
print_latex(c)
c
```

Nell'ellisse la distanza dai fuochi rimane sempre costante e questa costante $k=\overline{F_1X} + \overline{XF_2}$ e vale

$$
k = (c + a) + x
$$

```python
k = c+a+x
print_latex(k)
k
```

Per determinare $b$ consideriamo i triangoli equivalenti $F_1CO$ $F_2CO$

La distanza $d=\overline{F_1C}$ verifica la propriet√†

$$
\overline{F_1C} + \overline{F_1C} = k \;\implies\; 2d = k \;\implies d={k\over 2}
$$

```python
d = k/2
print_latex(d)
d
```

Per determinare $b$ utilizzo Pitagora

$$b = \sqrt{d^2-c^2}$$

```python
b = sqrt(d**2-c**2)
print_latex(b)
b
```
