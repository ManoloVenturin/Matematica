```python
# Toolbox simbolico
from sympy import *
```

```python
x= symbols('x', real=True)
a,b,c,d,e = symbols('a,b,c,d,e', real=True)
```

## Soluzione 1

Parto dal polinomio e applico le condizioni

$$\begin{cases}
P(0) = 0\\
P'(0) = 0\\
P(1) = 0\\
P(2) = -2\\
P'(2) = 0
\end{cases}$$

```python
P = a*x**4 + b*x**3 + c*x**2 + d*x + e
display(P)
P_prime = diff(P,x)
display(P_prime)
```

```python
print_latex(P)
print_latex(P_prime)
```

```python
# P(0)=0
eq1 = Eq(P.subs(x,0),0)
eq1
```

```python
# P'(0)=0
eq2 = Eq(P_prime.subs(x,0),0)
eq2
```

```python
# P(1)=0
eq3 = Eq(P.subs(x,1),0)
eq3
```

```python
# P(2)=-2
eq4 = Eq(P.subs(x,2),-2)
eq4
```

```python
# P'(2)=0
eq5 = Eq(P_prime.subs(x,2),0)
eq5
```

```python
sol = solve([eq1, eq2, eq3, eq4, eq5],[a,b,c,d,e])
sol
```

```python
# Polinomio
Popt = P.subs(sol)
Popt
```

```python
P_prime_opt = factor(P_prime.subs(sol))
P_prime_opt
```

```python
p1 = plot(Popt,(x,-1,3), ylim=(-5, 5), markers=[{'args': [[0,1,2],[0,0,-2], 'ro']}])
```

## Soluzione 2

$$\begin{cases}
P(0) = 0\\
P'(0) = 0\\
P(1) = 0\\
P(2) = -2\\
P'(2) = 0
\end{cases}$$

Parto dalla derivata prima che deve essere del tipo

$$P'(x) = x(x-2)(ax+b)$$

polinomio di terzo grado


```python
P2_prime = x*(x-2)*(a*x+b)
display(expand(P2_prime))
P2_prime
```

```python
print_latex(expand(P2_prime))
```

```python
# Ottengo polinomio come integrale con aggiunta della costante di integrazione
P2 = integrate(P2_prime,x) + c
P2
```

```python
print_latex(expand(P2))
```

```python
# Condizione P(0)=0
eq2a = Eq(P2.subs(x,0),0)
eq2a
```

```python
# Equazione P(1)=0
eq2b = Eq(P2.subs(x,1),0)
eq2b
```

```python
# Equazione P(2)=-2
eq2c = Eq(P2.subs(x,2),-2)
eq2c
```

```python
sol2 = solve([eq2a, eq2b, eq2c],[a,b,c])
sol2
```

```python
P2.subs(sol2)
```
